```python
import numpy as np
import pandas as pd

df = pd.read_csv('movie_data.txt', delimiter = '\t', header = None)
df.columns = ["ID", "title", "year", "rating", "votes", "duration", "genres"]  # columns named accordingly
df.set_index('ID')
```

```python
df['duration'] = df['duration'].apply(lambda x: int(x.replace('mins.','')))   # removes the minute
df['duration'] = df['duration']*60
df
```

```python
#df.info('title')
df['title'] = df['title'].apply(lambda x: (x[:-6]))
df
```

```python
movie_crew_df = pd.read_csv('crew_data.tsv.gz', delimiter = '\t', header = None, usecols=[0,1])
movie_crew_df.columns = ["ID", "director_ids"]  # columns named accordingly
movie_crew_df.set_index('ID')
```

```python
person_df = pd.read_csv('name_data.tsv.gz', delimiter = '\t', header = None)
person_df.shape
person_df = person_df.drop(person_df.columns[[2,3,4,5]], axis=1)
#person_df = person_df.iloc[,:4]
person_df.columns = ["person_ID", "name"]  # columns named accordingly
person_df.set_index('person_ID')
person_df.drop([0],axis=0)
person_df
```

Merge movie_crew_df into df with IDs and drop the columns without necessary data (hint outer-join)

```python
df = pd.merge(df,movie_crew_df,how='inner', on = 'ID')
df
```

Remove movies with more than one director and rename director_ids => director_id

```python
df = df[(df.director_ids).str.len()==9]
df
```

```python
df = df.rename(columns={'director_ids':'director_id'})
df
```

Merge df with person_df on director_id, remove director_id and person_ID columns, rename name => director

```python
person_df
```

```python
df = pd.merge(df, person_df, how = 'inner', left_on='director_id',right_on='person_ID')
df = df.rename(columns={'name':'director'})
df
```
